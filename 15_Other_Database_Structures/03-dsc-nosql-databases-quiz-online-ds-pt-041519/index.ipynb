## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

SQL databases are relational, NoSQL are non-relational, or distributed. SQL databases are table based and have a fixed schema. NoSQL are document, key-value pairs, graph or other based. They have a dynamic schema for unstructured data. Which is better depends on the circumstances. SQL are best for smaller DB, complex queries, standard data. NoSQL are better for unstructured data, and Big Data as they can be scaled horizontally across many different servers running concurrently.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MongoDB is a document store form of DB. It stores records as documents, rather than tables and rows. It is good for blogs, chat logs, mobile and social networking; it is not so good for highly transactional systems, or where data model is predetermined.







## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance is the ability to query the entire dataset, even if a server fails. RDD splits the data processing across multiple servers or nodes. If at any point one of the nodes fails, another node has a back-up of the data and can resume processing with no data loss.






## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce works by splitting big data into smaller chunks to process in parallel across multiple servers. The Map step determines how to split the data and which servers to use. The Reduce step aggregates the results to return a consolidated output.
